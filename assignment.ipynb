## Task 1 — Variables and Arithmetic (4 points)

Write a function `circle_stats(radius)` that takes a circle's radius as a float and returns a **tuple** `(circumference, area)` rounded to 2 decimal places.

Use: `circumference = 2 * π * r` and `area = π * r²`  
Use `import math` and `math.pi` for π.

**Example:**
```python
circle_stats(5)    # → (31.42, 78.54)
circle_stats(1)    # → (6.28, 3.14)
circle_stats(0)    # → (0.0, 0.0)
```

````python
import math

def circle_stats(radius):
    circumference = round(2 * math.pi * radius, 2)
    area = round(math.pi * (radius ** 2), 2)
    return (circumference, area)
````

````python
# VISIBLE TEST — Task 1
assert circle_stats(5) == (31.42, 78.54), f"Got: {circle_stats(5)}"
assert circle_stats(1) == (6.28, 3.14),   f"Got: {circle_stats(1)}"
assert circle_stats(0) == (0.0, 0.0),     f"Got: {circle_stats(0)}"
print("Task 1 visible tests passed.")
````

## Task 2 — String Operations (4 points)

Write a function `format_name(first, last)` that:
- Strips leading/trailing whitespace from both inputs
- Returns the name formatted as `"Last, First"` with each part capitalised (first letter upper, rest lower)

**Example:**
```python
format_name('anna', 'SMITH')       # → 'Smith, Anna'
format_name('  john  ', ' doe ')   # → 'Doe, John'
format_name('ALICE', 'wonderland') # → 'Wonderland, Alice'
```

````python
def format_name(first, last):
    clean_last = last.strip().capitalize()
    clean_first = first.strip().capitalize()
    return f"{clean_last}, {clean_first}"
````

````python
# VISIBLE TEST — Task 2
assert format_name('anna', 'SMITH') == 'Smith, Anna'
assert format_name('  john  ', ' doe ') == 'Doe, John'
assert format_name('ALICE', 'wonderland') == 'Wonderland, Alice'
print("Task 2 visible tests passed.")
````

## Task 3 — Boolean Logic (4 points)

Write a function `classify_triangle(a, b, c)` that takes three side lengths and returns a string:
- `'equilateral'` — all three sides equal
- `'isosceles'`  — exactly two sides equal
- `'scalene'`    — all sides different
- `'invalid'`    — not a valid triangle (any side ≤ 0, or triangle inequality violated)

Triangle inequality: each side must be less than the sum of the other two.

**Example:**
```python
classify_triangle(3, 3, 3)  # → 'equilateral'
classify_triangle(3, 3, 4)  # → 'isosceles'
classify_triangle(3, 4, 5)  # → 'scalene'
classify_triangle(1, 2, 10) # → 'invalid'
```

````python
def classify_triangle(a, b, c):
    # Check for valid side lengths
    if a <= 0 or b <= 0 or c <= 0:
        return 'invalid'

    # Check triangle inequality
    if (a + b <= c) or (a + c <= b) or (b + c <= a):
        return 'invalid'

    # Classify type
    if a == b == c:
        return 'equilateral'
    elif a == b or b == c or a == c:
        return 'isosceles'
    else:
        return 'scalene'
````

````python
# VISIBLE TEST — Task 3
assert classify_triangle(3, 3, 3) == 'equilateral'
assert classify_triangle(3, 3, 4) == 'isosceles'
assert classify_triangle(3, 4, 5) == 'scalene'
assert classify_triangle(1, 2, 10) == 'invalid'
assert classify_triangle(0, 3, 3) == 'invalid'
print("Task 3 visible tests passed.")
````
